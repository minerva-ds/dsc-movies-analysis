```python
import pandas as pd
import os

# Load the "in progress" pickle file
in_progress_pickle = 'in_progress.pkl'

# Check if the pickle file exists
if os.path.exists(in_progress_pickle):
    # Load the list of dictionaries
    with open(in_progress_pickle, 'rb') as f:
        in_progress_list = pd.read_pickle(f)
    
    # Convert the list of dictionaries to a DataFrame
    in_progress_df = pd.DataFrame(in_progress_list)
    
    # Display the first few rows to check the progress
    print(in_progress_df.head())
else:
    print("In-progress pickle file not found.")
```
